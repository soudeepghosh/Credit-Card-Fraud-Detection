Handling Imbalanced Datasets
1) Resampling(Undersampling and Oversampling)
2) SMOTE(Synthetic Minority Oversampling Technique)

 OverSampling

```python
# OverSampling
credit_card_data.Class.value_counts()
```

```python
from sklearn.utils import resample
#create two different dataframe of majority and minority class
df_majority = credit_card_data[(credit_card_data['Class']==0)]
df_minority = credit_card_data[(credit_card_data['Class']==1)]

# upsample minority class
df_minority_oversampled = resample(df_minority,
                                 replace=True,
                                 n_samples=284315,
                                 random_state=42)

# Combine majority class with upsampled minority class
df_oversampled = pd.concat([df_minority_oversampled, df_majority], axis=0)
df_oversampled.Class.value_counts()
```

Final Sampled Dataset:

```python
X_oversampled = df_oversampled.drop('Class', axis=1)
Y_oversampled = df_oversampled['Class']
X_oversampled.shape, Y_oversampled.shape
```

```python
count_classes = pd.value_counts(df_oversampled['Class'], sort = True)

count_classes.plot(kind = 'bar', rot=0, color='pink')

plt.title("After oversampling")

plt.xticks(range(2), LABELS)

plt.xlabel("Class")

plt.ylabel("Frequency")
```

Undersampling

```python
x = credit_card_data.drop('Class', axis=1)
y = credit_card_data['Class']
x.shape, y.shape
```

```python
# Undersampling
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

rus = RandomUnderSampler(random_state=42)

x_undersampled, y_undersampled = rus.fit_resample(x, y)
x_undersampled = pd.DataFrame(x_undersampled)
y_undersampled = pd.DataFrame(y_undersampled)
print(f"The number of Classes before the fit {Counter(y)}")
print(f"The number of Classes after the fit {Counter(y_undersampled)}")
```

```python
# Combining X_undersampled with y_undersampled
df_undersampled = pd.concat([x_undersampled, y_undersampled], axis=0)
df_undersampled.Class.value_counts()
```

```python
count_classes = pd.value_counts(df_undersampled['Class'], sort = True)

count_classes.plot(kind = 'bar', rot=0, color='green')

plt.title("After Undersampling")

plt.xticks(range(2), LABELS)

plt.xlabel("Class")

plt.ylabel("Frequency")
```

SMOTE(Synthetic Minority Oversampling Technique)

```python
# SMOTE(Synthetic Minority Oversampling Technique)
from imblearn.over_sampling import SMOTE
# Resampling the minority class. The strategy can be changed as required.
sm = SMOTE(sampling_strategy='minority', random_state=42)
# Fit the model to generate the data.
X_smote, y_smote = sm.fit_resample(x, y)
X_smote.shape
```

```python
df_smote = pd.concat([pd.DataFrame(X_smote), pd.DataFrame(y_smote)], axis=1)
df_smote.Class.value_counts()
```

```python
count_classes = pd.value_counts(df_smote['Class'], sort = True)

count_classes.plot(kind = 'bar', rot=0, color = 'red')

plt.title("After SMOTE (Synthetic Minority Oversampling Technique)")


plt.xticks(range(2), LABELS)

plt.xlabel("Class")

plt.ylabel("Frequency")
```
